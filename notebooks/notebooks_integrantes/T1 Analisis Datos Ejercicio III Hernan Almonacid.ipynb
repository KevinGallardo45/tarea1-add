```python
#carga datos
d2019=pd.read_csv("d2019.csv",index_col=0)
d2020=pd.read_csv("d2020.txt",index_col=0,sep=" ")
```

```python
df_sii_con=pd.concat((d2019,d2020),axis=0)
df_sii_con.head()
```

```python
#melt o unpivot de la tabla
df_sii=pd.melt(df_sii_con,id_vars=['Día','year'],var_name ='mes').iloc[:,[1,2,0,3]]
df_sii.columns=["year","mes","dia","valor"]
```

```python
df_sii.head(5)
```

```python
#guardar como csv
df_sii.to_csv("df_sii.csv")
```

#### Pregunta 3

##### 
Respecto a los valores nulos son días feriados y fines de semana. En dichos días no se cotiza el dolar.

```python
#Eliminación de fechas inexistentes en calendario
#Febrero
df_sii=df_sii.drop(df_sii[(df_sii["mes"]=="Feb")&(df_sii["dia"]>=29)&(df_sii["year"]==2019)].index)
df_sii=df_sii.drop(df_sii[(df_sii["mes"]=="Feb")&(df_sii["dia"]>=30)&(df_sii["year"]==2020)].index) #año bisiesto
#Abril
df_sii=df_sii.drop(df_sii[(df_sii["mes"]=="Abr")&(df_sii["dia"]>=31)&(df_sii["year"]==2019)].index)
df_sii=df_sii.drop(df_sii[(df_sii["mes"]=="Abr")&(df_sii["dia"]>=31)&(df_sii["year"]==2020)].index)
#Junio
df_sii=df_sii.drop(df_sii[(df_sii["mes"]=="Jun")&(df_sii["dia"]>=31)&(df_sii["year"]==2019)].index)
df_sii=df_sii.drop(df_sii[(df_sii["mes"]=="Jun")&(df_sii["dia"]>=31)&(df_sii["year"]==2020)].index)
#Septiembre
df_sii=df_sii.drop(df_sii[(df_sii["mes"]=="Sep")&(df_sii["dia"]>=31)&(df_sii["year"]==2019)].index)
df_sii=df_sii.drop(df_sii[(df_sii["mes"]=="Sep")&(df_sii["dia"]>=31)&(df_sii["year"]==2020)].index)
#Noviembre
df_sii=df_sii.drop(df_sii[(df_sii["mes"]=="Nov")&(df_sii["dia"]>=31)&(df_sii["year"]==2019)].index)
df_sii=df_sii.drop(df_sii[(df_sii["mes"]=="Nov")&(df_sii["dia"]>=31)&(df_sii["year"]==2020)].index)
```

```python
#creacion campos month y day

df_sii["month"] = np.where(df_sii['mes']=="Ene", 1,
                   np.where(df_sii['mes']=="Feb", 2,
                   np.where(df_sii['mes']=="Mar", 3,
                   np.where(df_sii['mes']=="Abr", 4,
                   np.where(df_sii['mes']=="May", 5,
                    np.where(df_sii['mes']=="Jun", 6,
                    np.where(df_sii['mes']=="Jul", 7,
                    np.where(df_sii['mes']=="Ago", 8,
                    np.where(df_sii['mes']=="Sep", 9,
                    np.where(df_sii['mes']=="Oct", 10,
                    np.where(df_sii['mes']=="Nov", 11,12)))))))))))    

df_sii["day"]=df_sii["dia"]
```

```python
#creacion campo fechas
df_sii["fecha"]=pd.to_datetime(df_sii[['year', 'month', 'day']])
```

```python
#creacion campo dia_de_la_semana
df_sii["dia_de_la_semana"]=df_sii['fecha'].dt.day_name()
df_sii["dia_de_la_semana"]=np.where(df_sii["dia_de_la_semana"]=="Monday","Lunes",
                           np.where(df_sii["dia_de_la_semana"]=="Tuesday","Martes",
                           np.where(df_sii["dia_de_la_semana"]=="Wednesday","Miercoles",
                           np.where(df_sii["dia_de_la_semana"]=="Thursday","Jueves",
                           np.where(df_sii["dia_de_la_semana"]=="Friday","Viernes",
                           np.where(df_sii["dia_de_la_semana"]=="Saturday","Sabado","Domingo"))))))

df_sii=df_sii[["year","mes","dia","valor","dia_de_la_semana"]]
df_sii
```

#### Pregunta 4

```python
#creación de un df para cada hoja del archivo excel
dolar_observado_bc = pd.read_excel("https://raw.githubusercontent.com/hfas85/tarea1-add/master/data/raw/dolar_observado_bc.xlsx",
                                 skiprows=2, dtype={'Periodo':str }, sheet_name=None)

df_1980 = dolar_observado_bc.get('80')
df_1990 = dolar_observado_bc.get('90')
df_2000 = dolar_observado_bc.get('00')
df_2010 = dolar_observado_bc.get('10')
df_2020 = dolar_observado_bc.get('20')

df_1980["Periodo"]=df_1980["Periodo"].str[0:10]
df_1990["Periodo"]=df_1990["Periodo"].str[0:10]
df_2000["Periodo"]=df_2000["Periodo"].str[0:10]
df_2010["Periodo"]=df_2010["Periodo"].str[0:10]
df_2020["Periodo"]=df_2020["Periodo"].str[0:10]
```

```python
#creación de campos year,mes,dia para cada df
df_1980[["year", "mes", "dia"]] = df_1980["Periodo"].str.split("-", expand = True)
df_1990[["year", "mes", "dia"]] = df_1990["Periodo"].str.split("-", expand = True)
df_2000[["year", "mes", "dia"]] = df_2000["Periodo"].str.split("-", expand = True)
df_2010[["year", "mes", "dia"]] = df_2010["Periodo"].str.split("-", expand = True)
df_2020[["year", "mes", "dia"]] = df_2020["Periodo"].str.split("-", expand = True)

df_1980.columns = ["Periodo","Dolar","year","mes","dia"]
df_1990.columns = ["Periodo","Dolar","year","mes","dia"]
df_2000.columns = ["Periodo","Dolar","year","mes","dia"]
df_2010.columns = ["Periodo","Dolar","year","mes","dia"]
df_2020.columns = ["Periodo","Dolar","year","mes","dia"]

df_2020.head()
```

```python
#creacion de un solo data frame
df_bc_all=pd.concat((df_1980,df_1990,df_2000,df_2010,df_2020),axis=0)
df_bc_all
```

```python
# Creación de csv agrupando por "year"
for (year), group in df_bc_all.groupby(['year']):
   group.to_csv(f'{year}.csv', index=False)
```
