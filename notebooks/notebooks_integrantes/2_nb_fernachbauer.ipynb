Concatene con pd.concat de dos formas:

> dolar_2019 y dolar_2020_txt solo con el argumento axis=0 guardandolo en el dataframe df_concat_0

¿Qué problema podríamos tener con las columnas en df_concat_1 y en los índices en df_concat_0?

El problema que se genera para df_concat_1 es que al ser un concat horizontal (merge), hay duplicidad de las columnas y se perdería la consistencia temporal de los datos.

El problema con df_concat_0 es que al ser una concatenación vertical (union), es posible que se pierda el correlativo de índices para los datos.
