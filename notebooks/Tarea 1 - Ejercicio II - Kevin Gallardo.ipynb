## Ejercicio II: Excel y tidy data:

```python
# Importando librerías
import pandas as pd
import os
```

#### II.1 Importando el archivo dolar2019_sii.csv
Al ser `;` el separador del csv es necesario usar el argumento `sep=";"`

```python
dolar_2019 = pd.read_csv("https://raw.githubusercontent.com/patoram123/tarea1-add/master/data/raw/dolar2019_sii.csv", sep=";")
dolar_2019.head()
```

#### II.2 Cargando dolar_2020_sii.txt

El archivo tiene barras como separador y además no trae el nombre de las columnas.
- Para el separador usaremos `sep="|"`
- Podemos extraer el nombre de las columnas desde el Dataframe anterior y usar `name=[]`
- Además, debemos usar `header=None`

```python
columns1=dolar_2019.columns
dolar_2020_txt = pd.read_csv("https://raw.githubusercontent.com/patoram123/tarea1-add/master/data/raw/dolar_2020_sii.txt", \
                             sep="|",
                            names=columns1, header=None)
dolar_2020_txt.head()
```

#### II.3 Cargando dolar_2020_sii_AA.txt
El separador en este caso se compone de 3 letras que parecen azarosas.
- Se usará regex para reconocer el patrón de 3 letras `sep="[a-zA-Zá-ý]{3}", engine='python'`
- `on_bad_lines="skip"` nos permitirá quitar la última línea que nos genera error

```python
columns1=dolar_2019.columns
dolar_2020_regex=pd.read_csv("https://raw.githubusercontent.com/patoram123/tarea1-add/master/data/raw/dolar_2020_sii_AA.dat",
                              sep="[a-zA-Zá-ý]{3}", engine='python',header=None, on_bad_lines="skip",
                            names=columns1)

dolar_2020_regex.head()
```

#### II.4 Wrangling y limpieza

En este caso las columnas entre los Dataframes se llamarán igual porque se han usado los nombres del primer Dataframe para leer los siguientes

```python
# Quitando última fila
dolar_2019.drop(31, inplace=True)
dolar_2020_txt.drop(31, inplace=True)

dolar_2020_txt.tail()
```

```python
# Agregando columna año

dolar_2019["year"] = 2019
dolar_2020_txt ["year"] = 2020
dolar_2020_regex ["year"] = 2020
```

```python
# Reemplazando comas por puntos

dolar_2019.replace(',', '.', regex=True, inplace=True)
dolar_2020_txt.replace(',', '.', regex=True, inplace=True)
dolar_2020_regex.replace(',', '.', regex=True, inplace=True)
```

```python
# Pasando todo a número (float)

dolar_2019 = dolar_2019.astype("float64")
print(dolar_2019.info())

dolar_2020_txt = dolar_2020_txt.astype("float64")
print(dolar_2020_txt.info())

dolar_2020_txt = dolar_2020_txt.astype("float64")
print(dolar_2020_txt.info())
```

#### II.5 Exportando

```python
#transformar dfs a archivos planos
dolar_2019.to_csv("..\data\interim\d2019.csv",sep=",")
dolar_2020_txt.to_csv("..\data\interim\..\data\interim\d2020.txt",sep=" ")
dolar_2020_regex.to_csv("..\data\interim\dolar_2020_excel.csv",sep=";")
```

#### II.6 Wrangling y limpieza

- El problema de usar solo `axis=1` es que se repiten las columnas sin combinarlas entre los dos datasets
- El problema de usar solo `axis=0` es que se repiten los indices, aunque en este caso sería facil solucionarlo usando `reset_index`

```python
df_concat_1= pd.concat((dolar_2019,dolar_2020_txt),axis=1)
df_concat_0= pd.concat((dolar_2019,dolar_2020_txt),axis=0)
```
