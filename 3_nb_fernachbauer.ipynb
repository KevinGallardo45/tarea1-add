1. En el notebook 3_nb_***.ipynb, cargue los dos archivos d2019.csv y d2020.txt en dos dataframes llamados df2019 y df2020 respectivamente.
```python
#Importamos los datos
d2019 = pd.read_csv("/content/d2019.csv")
d2020 = pd.read_csv("/content/d2020.txt", sep=' ')
```

```python
#Usamos el método melt
df_sii = pd.melt(d2019,
                 id_vars=['Día', 'year'],
                 value_vars=['Ene',	'Feb',	'Mar',	'Abr',	'May',	'Jun',	'Jul',	'Ago',	'Sep',	'Oct',	'Nov',	'Dic'],
                 value_name='valor',
                 var_name='Mes'
                 )
df_sii.head(100)
```

```python
#Guardamos csv
df_sii.to_csv('/content/df_sii.csv')
```

3. ¿Qué podrías inferir de los missing values y los días hábiles? Úsalos para crear una nueva variable llamada dia_de_la_semana en df_sii con valores "Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo".

* Los días en que el valor del dolar es missing, es porque el mercado está cerrado. Se podría interpretar que son los días hábiles.

```python
df_sii.groupby(['Día', 'Mes']).count()
```

```python
#Creamos una listra con los días de la semana
weekdays = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]
#Creamos una lista vacía para la columna
dia_de_la_semana = []
#seteamos j=1, dado que weekdays[1] es 'Martes' que fue el primer día del 2019.
j=1
#llenamos la columna de acuerdo al largo del dataframe
for i in range(0,df_sii.shape[0]):
  #llenamos la lista con el valor actual de dia de la semana
  dia_de_la_semana.append(weekdays[j])
  #Pasamos al siguiente día de la semana
  j=j+1
  #Cuando ya recorrió toda la semana
  if j%7==0:
    #Reiniciamos el indice semanal 
    j=1

#Creamos la columna 'dia_de_la_semana'
df_sii['dia_de_la_semana'] = dia_de_la_semana
df_sii.head()
```
